```python
import datetime
import re


try:
    from transformers import pipeline
    llm = pipeline("text-generation", model="distilgpt2", device=-1)
except Exception:
    llm = None


def college_info_tool(user_input):

    if "course" in user_input or "branch" in user_input:
        return "CBIT offers B.Tech, M.Tech, and MCA programs across various engineering disciplines."
    if "placement" in user_input:
        return "CBIT has strong placement support with top companies visiting every year."
    if "nss" in user_input:
        return "NSS focuses on community service, social responsibility, and leadership development."
    if "club" in user_input:
        return "CBIT has technical, cultural, social, and open-source clubs for overall student development."
    if "cbit" in user_input:
        return "CBIT (Chaitanya Bharathi Institute of Technology) is a reputed engineering college in Hyderabad."
    return None


def calculator(expression):
    try:
        return f"The result is {eval(expression)}"
    except:
        return "Invalid calculation."


def time_tool():
    return datetime.datetime.now().strftime("Current date and time: %d-%m-%Y %H:%M:%S")


def ai_agent(user_input):
    user_input = user_input.lower()

    if re.search(r'\d+\s*[\+\-\*/]\s*\d+', user_input):
        expression = re.findall(r'[\d\+\-\*/\.]+', user_input)[0]
        return calculator(expression)

    if "time" in user_input or "date" in user_input:
        return time_tool()

    college_response = college_info_tool(user_input)
    if college_response:
        return college_response

    if llm:
        response = llm(user_input, max_length=60, num_return_sequences=1)
        return response[0]["generated_text"]
    else:
        return "LLM unavailable. Ask about CBIT, time, or calculations."


print("College AI Assistant (type 'exit' to stop)")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Agent: Goodbye")
        break
    print("Agent:", ai_agent(user_input))
```
